```python
import wfdb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the ECG record from PhysioNet
record_name = "0379_002_019_ECG"          # <patientID>_<seg>_<hr>_ECG
db_dir      = "i-care/2.0/training/0379"  # path *inside* PhysioNet
rec = wfdb.rdrecord(record_name, pn_dir=db_dir)
```

```python
# All metadata fields
metadata = vars(rec)
df = pd.DataFrame(
    list(metadata.items()),
    columns=['Field', 'Value']
)

print(df)
```

```python
# --- explore metadata -------------------------------------------------
print(f"Sampling rate: {rec.fs} Hz")
print(f"Available channels: {rec.sig_name}")
print(f"Number of samples: {rec.sig_len}")
```

```python
# Extaer senal ECG 
ecg = rec.p_signal[:, 0] 
# Vector de tiempo        
t   = np.arange(ecg.size) / rec.fs


#Graficar senal ECG
plt.figure(figsize=(12,4))
seconds_to_plot = 5
samples         = int(seconds_to_plot * rec.fs)

plt.plot(t[:samples], ecg[:samples])
plt.title(f"Paciente 0379   {record_name}")
plt.xlabel("Time [s]")
plt.ylabel("Amplitud (mV)")
plt.tight_layout()
plt.show()
```

```python

ecg = rec.p_signal[:, 0]         
t   = np.arange(ecg.size) / rec.fs


#plot
plt.figure(figsize=(12,12))
seconds_to_plot = 5
samples         = int(seconds_to_plot * rec.fs)

for i in range(5):
    plt.subplot(5, 1, i + 1)
    plt.plot(t[samples*i:(i+1)*samples], ecg[samples*i:(i+1)*samples])
    plt.title(f"Patient 0320  first {seconds_to_plot}s of {record_name}")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude (mV)")
    
plt.tight_layout()
plt.show()
```
